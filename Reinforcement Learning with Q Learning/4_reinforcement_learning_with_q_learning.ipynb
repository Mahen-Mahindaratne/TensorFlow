```python
import gym
import numpy as np
```

Let's dive in with a practical example using OpenAI Gym's "FrozenLake" environment:

```python
# Import FrozenLake environment
env = gym.make("FrozenLake-v1")
```

```python
# Define action space and observation space
action_space = env.action_space.n
state_space = env.observation_space.n
```

```python
# Hyperparameters
alpha = 0.1  # Learning rate
gamma = 0.9  # Discount factor
epsilon = 0.1  # Exploration rate
```

```python
# Initialize Q-table with zeros
Q = np.zeros((state_space, action_space))
```

```python
# Iterate for specified number of episodes
for episode in range(100000):
  # Initialize starting state
    state = env.reset()

  # Episode loop
    while True:

        # Choose action based on Q-table and exploration policy
        if np.random.rand() < epsilon:
            action = env.action_space.sample()  # Explore!
        else:
            action = np.argmax(Q[state])  # Exploit!

        # Take action and observe new state and reward
        new_state, reward, done, _ = env.step(action)

        # Update Q-table
        max_future_Q = np.max(Q[new_state])
        Q[state, action] += alpha * (reward + gamma * max_future_Q - Q[state, action])

        # Check if episode finished
        if done:
            break

        # Update state for next iteration
        state = new_state
```

```python
# Print final Q-table
print(Q)
```

**Running the code:**

This code trains the agent to navigate the FrozenLake environment efficiently. It starts with random exploration and gradually learns the optimal path, reflected in the updated Q-table.
