```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
# Load and preprocess text data
text = ["This movie was amazing!", "I hated this product."]
labels = [1, 0]  # Positive = 1, Negative = 0
labels = np.array(labels) # Convert the list to a NumPy array
```

```python
tokenizer = Tokenizer(num_words=1000)
tokenizer.fit_on_texts(text)
sequences = tokenizer.texts_to_sequences(text)
padded_sequences = pad_sequences(sequences, maxlen=20)
```

```python
# Build and train the LSTM model
model = Sequential()
model.add(Embedding(1000, 128, input_length=20))
model.add(LSTM(64))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(padded_sequences, labels, epochs=10)
```

```python
# Use the model for prediction
new_text = "This play was truly captivating."
new_sequence = tokenizer.texts_to_sequences([new_text])
new_sequence = pad_sequences(new_sequence, maxlen=20)
prediction = model.predict(new_sequence)[0][0]
if prediction > 0.5:
    print("Predicted sentiment: Positive")
else:
    print("Predicted sentiment: Negative")
```

**Note:** This is a simplified example. Real-world NLP tasks often involve more complex architectures, larger datasets, and advanced techniques like attention mechanisms.
